# **Merge, Concat & Join Methods**

```python
ds1 = pd.Series(['A', 'B', 'C', 'A', 'A', 'B', 'C'])
ds2 = pd.Series([10, 20, 30, 40, 50, 60, 70])
ds3 = pd.Series(['A', 'B', 'C', 'D', 'A', 'B'])
ds4 = pd.Series([1, 2, 3, 4, 5, 6])

df1 = pd.DataFrame({
    "Prod": ds1,
    "ProdPrice": ds2
})

df2 = pd.DataFrame({
    "Prod": ds3,
    # "ProdPrice": ds2,
    "ProdQuantity": ds4
})
```

```python
# merge()

mergedDf = pd.merge(df2, df1, on='Prod')
print(df1)
print(df2)
mergedDf
```

```python
# join() -
print(df1)
print(df2)

joinedDf = df1.join(df2.set_index("Prod"), on='Prod')
joinedDf
```

```python
# concat() -
newDf = pd.concat([df1, df2], axis=1)
print(df1)
print(df2)
newDf
```
