# **Interpolate in Pandas**

```python
import pandas as pd
import numpy as np
```

```python
# Date format in pandas - YYYY-MM-DD
# Time Format in pandas - HH:MM:SS
# Date-Time format in padnas - YYYY-MM-DD HH:MM:SS

dataset = {
    "timestamp": pd.date_range("2026-04-21 07:00", periods=5, freq="min"),
    "tempreature": [23.65, np.nan, 27.21, None, 30.98]
}

iotDf = pd.DataFrame(dataset)
iotDf
```

```python
# here, there are 2 missing values and these values can't be filled by the use of fillna method. why?
# ---> because if we are going to fill a deafult value it may cause some analysis lack through which the insights may vary.

# to agar ye nahi kar sakte to kya kare fir?
# here, in the situation of no-default values where we can't use any default value we'll use interpolation through which the nan or none
# or missing values will be filled through the mean value of the previous and the next value

# to perform interpolation in pandas we are using interpolate method -

iotDf["tempreature"].interpolate()
```
