```python
import pandas as pd
# Ways to create a pandas series -
# ser = pd.Series({"Cars": ['Mercedes Benz', "BMW S", "Defender", "Range Rover", "Maruti 800"], "carPrice": [3, 1, 5, 3, .01]})
# print(ser)

cars = ['Mercedes Benz', "BMW S", "Defender", "Range Rover", "Maruti 800"]
carPrice = [3, 1, 5, 3, .01]

seriesOfCars = pd.Series(cars, carPrice)
print(seriesOfCars)
```

```python
# Ways to create a Pandas DataFrame -
productName = ["Wireless Headphone", "Macbook M5", "Bottle", "Room Freshner", "Chair", "Smartphone"]
productPrice = ["3999", "159999", "59", "499", "699", "129999"]
productCategory = ["Electronics", "Electronics", "Crockery", "Appliances", "Appliances", "Electronics"]

prodDf = pd.DataFrame({
    "Name": productName,
    "Price": productPrice,
    "Category": productCategory
})

# print(prodDf)
prodDf
```
