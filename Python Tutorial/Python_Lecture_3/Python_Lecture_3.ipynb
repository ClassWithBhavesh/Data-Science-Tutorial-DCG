## **Practice Question - **



1.   Write a program to check whether a given number is positive, negative, or zero.



```python

try:
  userNum = int(input("Enter a number to check whether it is '+', '-' or zero : "))
  if userNum > 0:
    print(f"{userNum} is a positive integer!")
  elif userNum < 0:
    print(f"{userNum} is a negative integer!")
  else:
    print(f"{userNum} is zero!")

except ValueError:
  print("The provided value must be in integer. You've entered a non integer value!")

```

2.   Write a program that checks if a given number is even or odd.



```python
try:
  userInput = int(input("Enter a number : "))
  if userInput == 0:
    print(f"{userInput} is not either even nor odd number!")
  elif userInput%2 == 0:
    print(f"{userInput} is a Even Number!")
  else:
    print(f"{userInput} is a Odd Number!")
except ValueError:
  print("provided value can't be operate! please provide an integer values....")
```
