# Exception Handling in Python

```python
try:
  num1 = int(input("Enter first number : "))
  num2 = int(input("Enter second number : "))


except ValueError:
  print("Entered input is not a number, please enter a number!")

except ZeroDivisionError:
  print("Invalid Input. Entered Number is Zero and we can't divide anything by zero!")

else:
    result = num1 / num2
    print(result)

# try-except block

# try - jo bhi kaam karana ho kisme karayenge? try bolck me!
# except - jo code try nahi sambhal payega wo kon sambhalega bhai? except block sambhalega by exception handling
```

```python
try:
  num = int(input("Enter a Number : "))

  print(f"You Entered - {num}")

except ValueError:
  print("Entered input is not a number, please enter a number!")
```

```python
try:
  number = [10,20,30,50,60,90]
  print(number[11])

except IndexError:
  print("Warning : Index Is out of provided list's range. please provide index within the list's range")
```

```python
try:
  student_dict = {
      "name": "Suraj",
      "age": 92,
      "course": "Data Science"
  }

  print(student_dict["marks"])

except KeyError:
  print("Key not found in the dictionary!")
```

```python
try:
  sum = "10" + 5
  print(sum)

except TypeError:
  print("data-type except integer are not operatable!")
```
