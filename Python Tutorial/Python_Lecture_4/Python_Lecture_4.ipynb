# ***Nested-If in Python***

```python
# real-life example for nested if -
# we have to arrange the provided values in ascending and descending order

# for 3 variable -
# ascending order -
num1 = int(input("Enter first value : "))
num2 = int(input("Enter second value : "))
num3 = int(input("Enter third value : "))

if num1 < num2 and num1 < num3:
  if num2 < num3:
    print("num1, num2, num3")
  elif num3 < num2:
    print("num1, num3, num2")
  else:
    print("num2 and num3 are equal, please provide different values!")
elif num2 < num1 and num2 < num3:
  if num1 < num3:
    print("num2, num1, num3")
  elif num3 < num1:
    print("num2, num3, num1")
  else:
    print("num1 and num3 are equal, please provide different values!")
elif num3 < num1 and num3 < num2:
  if num1 < num2:
    print("num3, num1, num2")
  elif num2 < num1:
    print("num3, num2, num1")
  else:
    print("num1 and num2 are equal, please provide different values!")
else:
  print("all provided values are equal, please provide different values!")




# descending order -
num1 = int(input("Enter first value : "))
num2 = int(input("Enter second value : "))
num3 = int(input("Enter third value : "))

if num1 > num2 and num1 > num3:
  if num2 > num3:
    print("num1, num2, num3")
  elif num3 > num2:
    print("num1, num3, num2")
  else:
    print("num2 and num3 are equal, please provide different values!")
elif num2 > num1 and num2 > num3:
  if num1 > num3:
    print("num2, num1, num3")
  elif num3 > num1:
    print("num2, num3, num1")
  else:
    print("num1 and num3 are equal, please provide different values!")
elif num3 > num1 and num3 > num2:
  if num1 > num2:
    print("num3, num1, num2")
  elif num2 > num1:
    print("num3, num2, num1")
  else:
    print("num1 and num2 are equal, please provide different values!")
else:
  print("all provided values are equal, please provide different values!")
```
