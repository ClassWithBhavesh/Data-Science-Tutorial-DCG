```python
# 3. Function with return value -
student_name = input("Enter Student Name - ")
hn = int(input("Enter Hindi Marks - "))
eng = int(input("Enter eng Marks - "))
maths = int(input("Enter maths Marks - "))
sc = int(input("Enter sc Marks - "))
sst = int(input("Enter sst Marks - "))
t_marks = int(input("Enter Total Marks of the Exams - "))

def obt_marks(hn, eng, maths, sc, sst):
  obt_marks = hn + eng + maths + sc + sst
  return obt_marks

def percentage(ob_marks, t_marks):
  percentage = ob_marks/t_marks * 100
  return percentage


def grading(std_name, ob_marks, percentage, total_marks):
  if percentage > 90 and percentage <= 100:
    print(f"{std_name} has got {ob_marks} out of {total_marks} and got {percentage}% and secured A Grade!")
  elif percentage > 80 and percentage <= 90:
    print(f"{std_name} has got {ob_marks} out of {total_marks} and got {percentage}% and secured B Grade!")
  elif percentage > 70 and percentage <= 80:
    print(f"{std_name} has got {ob_marks} out of {total_marks} and got {percentage}% and secured C Grade!")
  elif percentage > 50 and percentage <= 70:
    print(f"{std_name} has got {ob_marks} out of {total_marks} and got {percentage}% and secured D Grade!")
  elif percentage >= 33 and percentage <= 50:
    print(f"{std_name} has got {ob_marks} out of {total_marks} and got {percentage}% and secured D Grade!")
  else:
    print(f"{std_name} failed!")



ob_marks = obt_marks(hn, eng, maths, sc, sst)
per = percentage(ob_marks, t_marks)

grading(student_name, ob_marks, per, t_marks)
```
