# ***Control-Flow Statments in Python***



*   while
*   for

- to operate a loop we have to follow the below three steps -
    - declaring a variable
    - apply condition (the end point of the loop)
    - increment or decrement of that variable



```python
# 1. while -
num = 1

while num <= 10:
  print(num)
  num += 1
```

```python
# 2. for -

# for num in range(start, end, skip/steps)
for num in range(1, 11):
  print(num, end=' ')
```
