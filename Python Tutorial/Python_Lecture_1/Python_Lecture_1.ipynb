# ***Operators in Python***


*   operators - they are symbols which are used to perform certain task or activity
*   operands  - they are the values on which we are going to perform operations
*   operation - it is a procedure in which we will perform some operation on two or more operands


# Types of Operators in Python -


1.   Arithmetic Operators - +, -, *, /, //, **, %
2.   Assignment Operators - +=, -=, *=, /=, //=, **=, %=, =
3.   Comparision Operators - <, >, <=, >=, ==, !=
4.   Logical Operators - AND, OR, NOT
5.   Membership Operators - in, not in
6.   Identity Operators - is, is not
7.   Bitwise Operators  - & (AND), | (OR), ^ (XOR)
