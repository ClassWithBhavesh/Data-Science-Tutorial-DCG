# **Methods to built a NumPy Array**

```python
import numpy as np

arr = np.array([1,2,3,4,5,6,9])
print(arr)
np.ndim(arr)
```

**preparing 2D array in NumPy**

```python
_2DArr = np.array([
    [10,20,30], [1,2,3]
])

print(_2DArr)
print(np.ndim(_2DArr))
print(np.shape(_2DArr))
```

***preparing 3D array in NumPy***

```python
# _3DArr = np.array([
#     [10,20,30], [1,2,3], [4,5,6]
# ])

# print(_3DArr)
# print(np.ndim(_3DArr))
# print(np.shape(_3DArr))



_3DArr = np.array([[
    [10,20,30], [1,2,3], [4,5,6]
]])

print(_3DArr)
print(np.ndim(_3DArr))
print(np.shape(_3DArr))


_3DArr1 = np.array([
    [[10,20,30], [40,50,60]],
    [[1,2,3], [4,5,6]]
])

print(_3DArr1)
print(np.ndim(_3DArr1))
print(np.shape(_3DArr1))
```

***Arange Method to build an NumPy array within a range***

```python
# rangeArr = np.arange(startValue, endValue (end-1), skipValue)

rangeArr = np.arange(3, 101, 3)
print(rangeArr)

reverseArangeArr = np.arange(99, 0, -3)
print(reverseArangeArr)
```

***Type-Casting in  NumPy Arrays***

```python
arr = np.array([10,100,1000,10000])
print(arr)
print(arr.dtype)

int32Arr = np.array(arr, dtype=np.int16)
print(int32Arr)
print(int32Arr.dtype)

floatArr = np.array(arr, dtype=np.float64)
print(floatArr)
print(floatArr.dtype)
```

**LinSpace Method**

```python
eqDivArr = np.linspace(0, 20, 5, dtype=np.int64)
print(eqDivArr)
```

**Reshape Method in NumPy**

```python
_1DArr = np.arange(1, 37)
print(_1DArr)
print(np.ndim(_1DArr))

# Now, I am going to covert this linear array into a 2D-Array
# for this, we have to use reshape name ka method

_2DArr = _1DArr.reshape(2, 18)
print(_2DArr)

_2DArr = _1DArr.reshape(18, 2)
print(_2DArr)

_2DArr = _1DArr.reshape(6, 6)
print(_2DArr)

_2DArr = _1DArr.reshape(4, 9)
print(_2DArr)



# Alright! now we'll convert this _1DArr into _3DArr

_3DArr = _1DArr.reshape(2,2,9)   # page - 2  |  rows - 2  |  columns - 9
print(_3DArr)

_3DArr = _1DArr.reshape(2,3,6)   # page - 2  |  rows - 3  |  columns - 6
print(_3DArr)

_3DArr = _1DArr.reshape(6,2,3)   # page - 6  |  rows - 2  |  columns - 3
print(_3DArr)
```
