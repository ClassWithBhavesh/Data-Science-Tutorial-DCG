## **Fancy Indexing**

```python
import numpy as np

arr = np.arange(1, 11)
print(arr)

print(arr[0])
print(arr[2])
print(arr[4])
print(arr[6])

# fancy indexing -
print(arr[[0,2,4,6]])
```

# **Boolean Masking**

```python
# to filter out the even elements of the given array we'll use boolean masking on the basis of certain condition I'll filter the even array
evenArr = arr[arr%2 == 0]
print(evenArr)

oddArr = arr[arr%2 != 0]
print(oddArr)

greatThan5 = arr[arr > 5]
smallTahn5 = arr[arr < 5]

print(greatThan5)
print(smallTahn5)


filteredArr = arr[(arr > 2) | (arr < 8)]    # &
print(filteredArr)
```
