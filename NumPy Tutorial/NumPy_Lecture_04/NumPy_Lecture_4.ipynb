# **Slicing a NumPy Array**

```python
li = [1,2,32,4,5,6]
# print(li)

# print(li[startIndex:endIndex])  # Slicing Syntax

print(li[1:4])
```

```python
import numpy as np

arr = np.array([1,2,3,4,5,6])
print(arr)

arr1 = arr[:]
print(arr1)

arr1[0] = 11

print(arr)
print(arr1)
```

```python
# numpy slicing in 2D-array -

arr = np.arange(1, 13).reshape(3, 4)
print(arr)

# arr[rows, columns]
# arr[startRowIndex:EndRowIndex, startColumnIndex:endColumnIndex]
print(arr[0:2, 0:2])
print(arr[:2, :2])

print(arr[1:, 2:])

print(arr[:2, 1:3])
```

```python
# 3D-array - slicing
arr = np.arange(1,13).reshape(2,2,3)

print(arr)

print(arr[:, 1:3,:])
```
