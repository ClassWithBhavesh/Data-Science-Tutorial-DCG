# **Vectorization in NumPy**

- performing some operations on entire array at once with out any python loops
- why we must use vectorization in NumPy?
  - Eliminated the python loops
  - uses optimized C-level implementations
  - Much Faster as compared to python loops
  - easy to understand (cleaner code hota hai!)
  - Increases the readability of the code

```python
arr2 = np.array([1,2,3,4])
arr1 = np.array([10,20,30,40])

arr = arr1 - arr2
print(arr)
```

```python
li1 = [1,2,3,4]
li2 = [10,20,30,40]
print(li1 + li2)
```

```python
arr = np.array([14,15,23,56,88,96])

arr1 = arr + 5
print(arr)
print(arr1)
```
